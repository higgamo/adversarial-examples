```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def get_fnames(directory, start=None, end=None):
    assert start is not None or end is not None
    fnames = []
    for filename in os.listdir(directory):
        is_valid = False
        if start is None:
            is_valid = filename.endswith(end)
        elif end is None:
            is_valid = filename.startswith(start)
        elif end is not None and start is not None:
            is_valid = filename.startswith(start) and filename.endswith(end)
        else:
            raise ValueError
        if(is_valid):
            fnames.append(filename)
    return fnames
```

```python
directory = "/home/higgamo/Desktop/dataset/"
accuracy_filenames = get_fnames(directory, start = "ago", end = "a.npy")
fig = plt.figure()
for filename in accuracy_filenames:
    data = np.load(directory + filename)
    plt.plot(data,label=filename)
fig.suptitle('AdagradOptimizer accuracy', fontsize=16)
plt.xlabel('Steps (1 = 20steps)', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.legend()
plt.show()
```

```python
directory = "/home/higgamo/Desktop/dataset/"
accuracy_filenames = get_fnames(directory, start = "gdo", end = "l.npy")
fig = plt.figure()
for filename in accuracy_filenames:
    data = np.load(directory + filename)
    plt.plot(data/data[0],label=filename)
fig.suptitle('GradientDescentOptimizer Loss', fontsize=16)
plt.xlabel('Steps (1 = 20steps)', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend()
plt.show()
```
